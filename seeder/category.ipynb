# Category Data Seeder
This notebook seeds 10 categories to MongoDB for RAG filtering.

```python
import sys
from pathlib import Path

# Add parent directory to path so we can import db module
parent_dir = str(Path.cwd().parent)
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

print(f'Added to path: {parent_dir}')
```

```python
from db import get_mongo_db

def seed_categories():
    """Seed 10 categories to MongoDB for RAG filtering"""
    db = get_mongo_db()

    categories = [
        {"name": "Wildlife", "is_active": True},
        {"name": "Geography", "is_active": True},
        {"name": "History", "is_active": True},
        {"name": "Science", "is_active": True},
        {"name": "Technology", "is_active": True},
        {"name": "Education", "is_active": True},
        {"name": "Health", "is_active": True},
        {"name": "Sports", "is_active": True},
        {"name": "Entertainment", "is_active": True},
        {"name": "Politics", "is_active": True},
        {"name": "Bibliography", "is_active": True},
        {"name": "Philosophy", "is_active": True},
        {"name": "Biography", "is_active": True},
        {"name": "Leave Policy", "is_active": True}
    ]

    # Clear existing categories (optional)
    db["category"].delete_many({})

    # Insert new categories
    result = db["category"].insert_many(categories)
    print(f"Seeded {len(result.inserted_ids)} categories")
    return result
```

```python
# Run the seeder
seed_categories()
```
